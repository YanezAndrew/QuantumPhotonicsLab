```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
import numpy as np
```

```python
df = pd.read_csv('data_TEST.csv')
df
```

```python
df['Voltage'] = df['Voltage'].apply(ast.literal_eval)
#df['Voltage'].values[0].apply(ast.literal_eval)
```

```python
df['Amps'] = df['Amps'].apply(ast.literal_eval)
```

```python

xvals = np.array(df['Voltage'].values[0])
yvals = np.array(df['Amps'].values[0])

xvals_last = np.array(df['Voltage'].values[0])
yvals_last = np.array(df['Amps'].values[-1])
```

```python


plt.figure(dpi=1500)
plt.semilogy(xvals, abs(yvals), color = 'red', label = '300K')
plt.semilogy(xvals_last, abs(yvals_last), color = 'darkblue', label = '4K')

plt.xlim(-5, 1)
plt.xlabel('Voltage [V]')
plt.ylabel('Current [Log(A)]')
plt.legend()
```

```python
plt.figure(dpi=1500)
plt.plot(xvals, 1000 * yvals, color = 'red', label = '300K')
plt.plot(xvals_last, 1000 * yvals_last, color = 'darkblue', label = '4K')

plt.xlim(0, 1.5)
plt.xlabel('Voltage [V]')
plt.ylabel('Current [mA)]')
plt.legend()
```
